#Solar Panel Energy efficiency

```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
X = train.drop(["efficiency", "id"], axis = 1)
Y = train["efficiency"]
```

```python
test.columns
```

```python
train.columns
```

#Data Preprocessing

```python
X
```

```python
Y
```

```python
X.isna().sum()
```

```python
X["cloud_coverage"] = X["cloud_coverage"].fillna(X["cloud_coverage"].mean())
cols = ["temperature", "irradiance", "panel_age", "soiling_ratio", "voltage", "current", "module_temperature"]
X[cols] = X[cols].fillna(X[cols].mean())
X["maintenance_count"] = X["maintenance_count"].fillna(X["maintenance_count"].mode()[0])
```

```python
X["error_code"] = X["error_code"].apply(
    lambda x : np.random.choice(X["error_code"].dropna()) if pd.isna(x) else x
)
X["installation_type"] = X["installation_type"].apply(
    lambda x : np.random.choice(X["installation_type"].dropna()) if pd.isna(x) else x
)
```

```python
X.head()
```

```python
X.info()
```

```python
X.isna().sum()
```

#Feature Scaling

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
encode = ["string_id" ,"error_code" ,"installation_type"]
X["string_id"] = le.fit_transform(X["string_id"])
X["error_code"] = le.fit_transform(X["error_code"])
X["installation_type"] = le.fit_transform(X["installation_type"])
```

```python
string_columns = X.select_dtypes(include=[ 'string']).columns.tolist()
print("String columns:", string_columns)
```

```python
Xerror = X.apply(pd.to_numeric, errors="coerce")
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_scaled = sc.fit_transform(Xerror)
```

```python
X_scaled
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xval, Ytrain, Yval = train_test_split(X_scaled, Y, test_size = 0.2, random_state = 2)
```

```python
Xtrain.shape
```

```python
Ytrain.shape
```

```python
Yval.shape
```

```python
print(np.any(np.isnan(Xtrain)))     # Should be False
print(np.any(np.isnan(Ytrain)))     # Should be False
print(np.any(np.isinf(Xtrain)))     # Should be False
print(np.any(np.isinf(Ytrain)))     # Should be False
print(np.any(np.isnan(Xval)))
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
Xtrain = imputer.fit_transform(Xtrain)
```

```python
Xval = imputer.fit_transform(Xval)
```

```python
from sklearn import metrics
from sklearn.ensemble import RandomForestRegressor, VotingRegressor
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV

param_grid_rf = {
    "n_estimators" : [100, 200],
    "max_depth" : [5, 10, None],
    "max_features" : [0.8],
    "min_weight_fraction_leaf" : [0.01],
    "min_samples_split" : [5, 10]
}

param_grid_xg = {
    "n_estimators" : [100, 200],
    "max_depth" : [3,6],
    "learning_rate" : [0.05, 0.1],
    "subsample" : [0.6, 0.8, 1],
    "colsample_bytree": [0.6, 0.8, 1]
}

rf = RandomForestRegressor()
xg = XGBRegressor()

grid_rf = GridSearchCV(rf, param_grid_rf, cv = 3, scoring="neg_mean_squared_error", n_jobs=-1)
grid_xg = GridSearchCV(xg, param_grid_xg, cv = 3, scoring="neg_mean_squared_error", n_jobs=-1)

grid_rf.fit(Xtrain, Ytrain)
grid_xg.fit(Xtrain, Ytrain)

best_rf = grid_rf.best_estimator_
best_xg = grid_xg.best_estimator_

voting_model = VotingRegressor(estimators=[("rf",best_rf), ("xg", best_xg)], n_jobs = 5)
voting_model.fit(Xtrain, Ytrain)
```

```python
Ypred = voting_model.predict(Xval)

100*(1-np.sqrt(metrics.mean_squared_error(Yval,Ypred)))
```
